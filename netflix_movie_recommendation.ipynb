```python
filtered_MovieList = movie[movie.revenue>20000]
filetered_MovieList = filtered_MovieList[filtered_MovieList.budget>15000]
filtered_MovieList = filtered_MovieList[filtered_MovieList.vote_count>100]
filtered_MovieList = filtered_MovieList[filtered_MovieList.vote_average>7.2]
```

```python
from sklearn import preprocessing

selected_columns = ['revenue', 'budget', 'vote_average', 'vote_count']
ScaledMovieData = filtered_MovieList[selected_columns].copy()

# Standardization using StandardScaler
Scaler = preprocessing.StandardScaler().fit(ScaledMovieData)
normalized_data = pd.DataFrame(Scaler.transform(ScaledMovieData), columns=ScaledMovieData.columns, index=ScaledMovieData.index)

print(normalized_data)
```

```python
filtered_MovieList['revenue'] = normalized_data['revenue']
filtered_MovieList['budget'] = normalized_data['budget']
filtered_MovieList['vote_average'] = normalized_data['vote_average']
filtered_MovieList['vote_count'] = normalized_data['vote_count']
filtered_MovieList['Rating'] = movie['vote_average']
```

```python
filtered_MovieList.head(10)
```

```python
filtered_MovieList['My_Recommendation'] = (filtered_MovieList['revenue']/filtered_MovieList['budget'])*filtered_MovieList['vote_average']*filtered_MovieList['vote_count']
```

```python
RecommendedMovies = filtered_MovieList.sort_values("My_Recommendation",ascending=False)[["original_title","popularity","My_Recommendation","Rating"]].head(20)
RecommendedMovies
```

```python
DatasetBasedReccomendation = filtered_MovieList.sort_values("popularity",ascending=False)[["original_title","popularity","vote_average"]].head(20)
DatasetBasedReccomendation

```

```python
# Convert lists to sets for efficient comparison
set1 = set(RecommendedMovies["original_title"])
set2 = set(DatasetBasedReccomendation["original_title"])

# Find common movies
common_movies = set1.intersection(set2)

# Display common movie names
print("Common Movies:")
for original_title in common_movies:
    print(original_title)
```
